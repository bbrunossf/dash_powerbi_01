```python
def gerar_obras_temporais(n, start_date, end_date):
    """
    Gera distribuição de obras simulando evolução temporal:
    - 2024: mais obras, menor duração cada (obras iniciando/finalizando)
    - 2025: menos obras, maior duração cada (obras em andamento)
    """
    n_obras_ativas = 110
    obras_ativas = np.random.choice(248, n_obras_ativas, replace=False) + 1
    
    # Dividir período em 2024 e 2025
    split_date = datetime(2025, 1, 1)
    total_days = (end_date - start_date).days
    days_2024 = (split_date - start_date).days
    days_2025 = (end_date - split_date).days
    
    # Proporção de registros por período (baseado nos dias)
    prop_2024 = days_2024 / total_days
    prop_2025 = days_2025 / total_days
    
    registros_2024 = int(n * prop_2024)
    registros_2025 = n - registros_2024
    
    # Configuração por período
    # 2024: mais obras ativas (80% das obras), registros distribuídos
    obras_ativas_2024 = int(n_obras_ativas * 0.8)  # 88 obras
    # 2025: menos obras ativas (50% das obras), mais concentrado
    obras_ativas_2025 = int(n_obras_ativas * 0.5)  # 55 obras
    
    # Selecionar obras para cada período
    obras_2024 = np.random.choice(obras_ativas, obras_ativas_2024, replace=False)
    obras_2025 = np.random.choice(obras_ativas, obras_ativas_2025, replace=False)
    
    # Distribuir registros 2024 (mais uniforme - obras menores)
    registros_por_obra_2024 = np.random.poisson(registros_2024 / obras_ativas_2024, obras_ativas_2024)
    registros_por_obra_2024 = np.maximum(registros_por_obra_2024, 1)  # Mínimo 1
    
    # Ajustar para soma exata
    diff_2024 = registros_2024 - registros_por_obra_2024.sum()
    if diff_2024 > 0:
        indices = np.random.choice(obras_ativas_2024, diff_2024, replace=True)
        for idx in indices:
            registros_por_obra_2024[idx] += 1
    elif diff_2024 < 0:
        for _ in range(-diff_2024):
            idx_max = np.argmax(registros_por_obra_2024)
            if registros_por_obra_2024[idx_max] > 1:
                registros_por_obra_2024[idx_max] -= 1
    
    # Distribuir registros 2025 (mais concentrado - obras maiores)
    # Usar distribuição exponencial para concentrar mais
    pesos_2025 = np.random.exponential(2, obras_ativas_2025)
    pesos_2025 = pesos_2025 / pesos_2025.sum()
    registros_por_obra_2025 = (pesos_2025 * registros_2025).astype(int)
    registros_por_obra_2025 = np.maximum(registros_por_obra_2025, 1)
    
    # Ajustar para soma exata
    diff_2025 = registros_2025 - registros_por_obra_2025.sum()
    if diff_2025 > 0:
        indices_maiores = np.argsort(registros_por_obra_2025)[-diff_2025:]
        registros_por_obra_2025[indices_maiores] += 1
    elif diff_2025 < 0:
        for _ in range(-diff_2025):
            idx_max = np.argmax(registros_por_obra_2025)
            if registros_por_obra_2025[idx_max] > 1:
                registros_por_obra_2025[idx_max] -= 1
    
    # Criar listas de registros por período
    registros_2024_list = []
    for i, obra_id in enumerate(obras_2024):
        count = registros_por_obra_2024[i]
        # Gerar timestamps para 2024
        timestamps = np.random.uniform(
            start_date.timestamp(), 
            split_date.timestamp(), 
            count
        )
        for ts in timestamps:
            registros_2024_list.append({
                'obra_id': obra_id,
                'timestamp': ts,
                'periodo': '2024'
            })
    
    registros_2025_list = []
    for i, obra_id in enumerate(obras_2025):
        count = registros_por_obra_2025[i]
        # Gerar timestamps para 2025
        timestamps = np.random.uniform(
            split_date.timestamp(), 
            end_date.timestamp(), 
            count
        )
        for ts in timestamps:
            registros_2025_list.append({
                'obra_id': obra_id,
                'timestamp': ts,
                'periodo': '2025'
            })
    
    # Combinar e embaralhar todos os registros
    todos_registros = registros_2024_list + registros_2025_list
    np.random.shuffle(todos_registros)
    
    return todos_registros[:n]
```

```python
def gerar_dados_ficticios_temporal(n):
    """
    Gera dados fictícios com evolução temporal das obras
    """
    registros = []
    
    # Configurações
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2025, 7, 30)
    
    # Proporções para tipos de tarefa
    tipos_tarefa = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    #proporcoes_tarefa = [0.146, 0.146, 0.146, 0.146, 0.146, 0.10, 0.12, 0.05, 0.02]
    proporcoes_tarefa = [0.146, 0.18, 0.146, 0.2, 0.146, 0.08, 0.032, 0.05, 0.02] #tem que somar 1
    
    # Gerar distribuição temporal de obras
    registros_temporais = gerar_obras_temporais(n, start_date, end_date)
    
    # Gerar outros campos
    id_tipo_tarefa_array = np.random.choice(tipos_tarefa, n, p=proporcoes_tarefa)
    id_nome_array = np.random.randint(1, 9, n)
    id_categoria_array = np.random.randint(1, 6, n)
    hora_extra_array = np.random.choice([True, False], n, p=[0.2, 0.8])
    
    # Horários comerciais
    horas_comerciais = np.random.randint(8, 20, n)
    minutos = np.random.randint(0, 60, n)
    segundos = np.random.randint(0, 60, n)
    
    for i in range(n):
        # Usar timestamp do registro temporal
        timestamp_base = datetime.fromtimestamp(registros_temporais[i]['timestamp'])
        timestamp = timestamp_base.replace(
            hour=horas_comerciais[i], 
            minute=minutos[i],
            second=segundos[i]
        )
        
        registro = {
            "timestamp": timestamp,
            "id_nome": int(id_nome_array[i]),
            "id_obra": int(registros_temporais[i]['obra_id']),
            "id_tipo_tarefa": int(id_tipo_tarefa_array[i]),
            "id_categoria": int(id_categoria_array[i]),
            "duracao_minutos": 60,  # Sempre 60 (1 hora por registro)
            "hora_extra": bool(hora_extra_array[i])
        }
        registros.append(registro)
    
    return registros

# Gerar 16000 registros
dados_ficticios = gerar_dados_ficticios_temporal(16000)

# Análise dos resultados
import pandas as pd
df = pd.DataFrame(dados_ficticios)
df['ano'] = df['timestamp'].dt.year
df['horas'] = df['duracao_minutos'] / 60

print("=== ANÁLISE TEMPORAL ===")
print("\nRegistros por ano:")
print(df['ano'].value_counts().sort_index())

print("\nHoras por obra em 2024 (top 10):")
horas_2024 = df[df['ano'] == 2024].groupby('id_obra')['horas'].sum().sort_values(ascending=False).head(10)
print(horas_2024)

print("\nHoras por obra em 2025 (top 10):")
horas_2025 = df[df['ano'] == 2025].groupby('id_obra')['horas'].sum().sort_values(ascending=False).head(10)
print(horas_2025)

print(f"\nObras ativas em 2024: {df[df['ano'] == 2024]['id_obra'].nunique()}")
print(f"Obras ativas em 2025: {df[df['ano'] == 2025]['id_obra'].nunique()}")
print(f"Total de obras ativas: {df['id_obra'].nunique()}")

df.to_csv("dados_temporais.csv", index=False)
print("\nArquivo 'dados_temporais.csv' salvo!")
```

```python
def criar_perfis_funcionarios():
    """
    Cria perfis de funcionários com especializações em diferentes tipos de tarefa
    """
    perfis = {
        # Perfil 1: Especialista em tarefas administrativas (tipos 1, 2)
        1: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0, 0.75, 0.1, 0.05, 0.05, 0.02, 0.02, 0.005, 0.005]
        },
        # Perfil 2: Técnico generalista (distribuição mais uniforme)
        2: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0.15, 0.15, 0.15, 0.15, 0.15, 0.08, 0.08, 0.045, 0.045]
        },
        # Perfil 3: Especialista em tarefas complexas (tipos 6, 7, 8)
        3: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0.25, 0.05, 0.1, 0.1, 0.1, 0.05, 0.25, 0.08, 0.02]
        },
        # Perfil 4: Especialista técnico avançado (tipos 8, 9)
        4: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0.02, 0.03, 0.05, 0.05, 0.1, 0.15, 0.15, 0.3, 0.15]
        },
        # Perfil 5: Operacional básico (tipos 1, 3, 4, 5)
        5: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0.25, 0.1, 0.25, 0.2, 0.15, 0.02, 0.02, 0.005, 0.005]
        },
        # Perfil 6: Coordenador (tipos 4, 5, 6, 7)
        6: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0.08, 0.08, 0.09, 0.2, 0.2, 0.15, 0.15, 0.03, 0.02]
        },
        # Perfil 7: Especialista em execução (tipos 3, 4, 5)
        7: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0.1, 0.1, 0.3, 0.25, 0.2, 0.02, 0.02, 0.005, 0.005]
        },
        # Perfil 8: Líder técnico (tipos 6, 7, 8, 9)
        8: {
            'tipos_tarefa': [1, 2, 3, 4, 5, 6, 7, 8, 9],
            'proporcoes': [0.03, 0.03, 0.04, 0.05, 0.1, 0.2, 0.25, 0.2, 0.1]
        }
    }
    return perfis

def gerar_dados_com_perfis_funcionarios(n):
    """
    Gera dados com perfis diferenciados por funcionário
    """
    registros = []
    
    # Configurações
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2025, 7, 30)
    perfis_funcionarios = criar_perfis_funcionarios()
    n_funcionarios = 8  # id_nome de 1 a 8
    
    # Atribuir perfil para cada funcionário
    funcionario_perfil = {}
    perfis_disponiveis = list(perfis_funcionarios.keys())
    for func_id in range(1, n_funcionarios + 1):
        funcionario_perfil[func_id] = perfis_disponiveis[func_id - 1]
    
    # Gerar distribuição temporal de obras
    registros_temporais = gerar_obras_temporais(n, start_date, end_date)
    
    # Gerar funcionários (distribuição relativamente uniforme)
    id_nome_array = np.random.randint(1, n_funcionarios + 1, n)
    
    # Gerar outros campos básicos
    id_categoria_array = np.random.randint(1, 6, n)
    hora_extra_array = np.random.choice([True, False], n, p=[0.2, 0.8])
    
    # Horários comerciais
    horas_comerciais = np.random.randint(8, 20, n)
    minutos = np.random.randint(0, 60, n)
    segundos = np.random.randint(0, 60, n)
    
    # Gerar tipos de tarefa baseados no perfil do funcionário
    id_tipo_tarefa_array = np.zeros(n, dtype=int)
    
    for i in range(n):
        funcionario_id = id_nome_array[i]
        perfil_id = funcionario_perfil[funcionario_id]
        perfil = perfis_funcionarios[perfil_id]
        
        # Gerar tipo de tarefa baseado no perfil do funcionário
        id_tipo_tarefa_array[i] = np.random.choice(
            perfil['tipos_tarefa'], 
            p=perfil['proporcoes']
        )
    
    for i in range(n):
        # Usar timestamp do registro temporal
        timestamp_base = datetime.fromtimestamp(registros_temporais[i]['timestamp'])
        timestamp = timestamp_base.replace(
            hour=horas_comerciais[i], 
            minute=minutos[i],
            second=segundos[i]
        )
        
        registro = {
            "timestamp": timestamp,
            "id_nome": int(id_nome_array[i]),
            "id_obra": int(registros_temporais[i]['obra_id']),
            "id_tipo_tarefa": int(id_tipo_tarefa_array[i]),
            "id_categoria": int(id_categoria_array[i]),
            "duracao_minutos": 60,
            "hora_extra": bool(hora_extra_array[i])
        }
        registros.append(registro)
    
    return registros, funcionario_perfil, perfis_funcionarios

# Gerar dados com perfis
dados_ficticios, funcionario_perfil, perfis = gerar_dados_com_perfis_funcionarios(24320)

# Análise dos perfis
import pandas as pd
df = pd.DataFrame(dados_ficticios)

print("=== ANÁLISE DE PERFIS DE FUNCIONÁRIOS ===")
print("\nPerfil de cada funcionário:")
for func_id, perfil_id in funcionario_perfil.items():
    print(f"Funcionário {func_id}: Perfil {perfil_id}")

print("\nDistribuição de tipos de tarefa por funcionário:")
pivot_funcionario = df.groupby(['id_nome', 'id_tipo_tarefa']).size().unstack(fill_value=0)
percentuais = pivot_funcionario.div(pivot_funcionario.sum(axis=1), axis=0) * 100

for func_id in range(1, 9):
    print(f"\nFuncionário {func_id}:")
    for tipo in range(1, 10):
        if tipo in percentuais.columns:
            pct = percentuais.loc[func_id, tipo] if func_id in percentuais.index else 0
            print(f"  Tipo {tipo}: {pct:.1f}%")

print("\nHoras totais por funcionário:")
horas_por_funcionario = df.groupby('id_nome')['duracao_minutos'].sum() / 60
print(horas_por_funcionario)

print("\nHoras totais por tipo de tarefa:")
horas_por_tipo = df.groupby('id_tipo_tarefa')['duracao_minutos'].sum() / 60
print(horas_por_tipo.sort_index())

df.to_csv("dados_com_perfis.csv", index=False)
print("\nArquivo 'dados_com_perfis.csv' salvo!")

# Salvar também os perfis para referência
import json
with open("perfis_funcionarios.json", "w") as f:
    perfis_info = {
        "funcionario_perfil": funcionario_perfil,
        "perfis_funcionarios": perfis
    }
    json.dump(perfis_info, f, indent=2)
print("Arquivo 'perfis_funcionarios.json' salvo!")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def simular_distribuicao_clientes():
    n_simulacoes = 10000
    n_obras_por_cliente = 10
    n_clientes = 25  # Assumindo 248 obras ÷ 10 = ~25 clientes
    prob_selecao = 110/248
    
    distribuicao_clientes = []
    
    for _ in range(n_simulacoes):
        obras_por_cliente = []
        for cliente in range(n_clientes):
            # Cada cliente tem 10 obras, cada uma com prob_selecao chance
            obras_selecionadas = np.random.binomial(n_obras_por_cliente, prob_selecao)
            obras_por_cliente.append(obras_selecionadas)
        distribuicao_clientes.append(obras_por_cliente)
    
    # Analisar resultados
    distribuicao_clientes = np.array(distribuicao_clientes)
    media_obras = np.mean(distribuicao_clientes, axis=0)
    std_obras = np.std(distribuicao_clientes, axis=0)
    
    print(f"Média de obras por cliente: {np.mean(media_obras):.2f}")
    print(f"Desvio padrão entre clientes: {np.std(media_obras):.2f}")
    print(f"Min-Max obras por cliente: {np.min(media_obras):.1f} - {np.max(media_obras):.1f}")
    
    return distribuicao_clientes

# Rodar simulação
distribuicao = simular_distribuicao_clientes()
```
